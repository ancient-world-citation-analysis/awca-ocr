```python
import requests
import pytesseract
from PIL import Image
url = 'https://www.researchgate.net/profile/Diane-Schallert/publication/288057676/figure/fig1/AS:613969366622247@1523393112783/Figure-A2-Island-text-Neutral-version-English.png'
src = Image.open(requests.get(url, stream=True).raw)
pytesseract.image_to_osd(src)
```

```python
pytesseract.image_to_osd(src, config=' --dpi 70 ')
```

```python
import fitz
def image(document, page_number, resolution=5):
  pix = document.get_page_pixmap(
      page_number, matrix=fitz.Matrix(resolution, resolution))
  return Image.frombytes(
      ("RGBA" if pix.alpha else "RGB"),
      [pix.width, pix.height], pix.samples
      )
handbook = fitz.open('/content/drive/MyDrive/AWCA/Colab_notebooks/OCR/MixLa'
                     'ngTest/Joseph-Janda--Handbook of Historical Linguisti'
                     'cs--Bibliography (OCR).pdf')
image = image(handbook, 0, 1)
image
```

```python
pytesseract.image_to_string(image)
```

```python
print(pytesseract.image_to_data(image.resize((image.width*2, image.height*2))))
```

```python
print(pytesseract.image_to_string(image.resize((image.width*2, image.height*2))))
```
